## Helper functions

```python
def generate_singed_data_stream(length):
    """
        generate seasonal sin  waves data stream with noises
    """
    
def generate_seasonal_data_with_anomalies(length, seasonality_amplitude=1, anomaly_rate=0.05, anomaly_duration=5, anomaly_magnitude=5):
        """
        Generates a seasonal data series with random anomalies.
        
        Args:
        length: Length of the data series.
        seasonality_amplitude: Amplitude of the seasonal component.
        anomaly_rate: Proportion of data points that will be anomalies.
        anomaly_duration: Duration of each anomaly in days.
        anomaly_magnitude: Magnitude of the anomalies.

        Returns:
        A pandas Series representing the generated data.
        """

        # Generate time index
        time_index = pd.date_range(start='2023-01-01', periods=length, freq='D')

        # Generate seasonal component
        seasonal = seasonality_amplitude * np.sin(2 * np.pi * np.arange(length) / 365)

        # Generate random noise
        noise = np.random.randn(length)

        # Generate anomalies
        anomaly_indices = np.random.choice(length, int(length * anomaly_rate), replace=False)

        anomalies = np.zeros(length)
        for index in anomaly_indices:
            start_index = max(0, index - anomaly_duration // 2)
            end_index = min(length - 1, index + anomaly_duration // 2)
            anomalies[start_index:end_index + 1] = anomaly_magnitude

        # Combine components
        data = seasonal + noise + anomalies

        return pd.Series(data, index=time_index)
    
def mean_abs_devision(data):
    """
        apply MAD algorithms od data
    """
    df= pd.DataFrame(data, columns=["values"])
    model = MAD().fit(df)
    labels = model.labels_
    return labels

def isolate_forest(data, combinations = 0.05):
    """
        apply isolate forests algorithm on data
    """
    model = IsolationForest(contamination=combinations)
    model.fit(data.values.reshape(-1,1))
    
    # Predict anomalies
    predictions = model.predict(data.values.reshape(-1,1))
    return predictions
    
def Generat_datae_with_noises(sample_size, upper_bound, anomaly_rate) :
    """
    Return nd-array of float points
    :param sample_size: length of sequence
    :param upper_bound: the upper bound of the sequence values
    :return:
    """
    indices = pd.date_range(start='1-1-2024', end='30-1-2024', periods=sample_size)
    data = np.random.random(sample_size) * upper_bound
    anomaly_indices = np.random.choice(sample_size, int(sample_size * anomaly_rate), replace=False)

    data[anomaly_indices] *= 5  # Introduce anomalies as outliers

    return pd.Series(data = data, index= indices)
```

# Random Data Stream 


## Test Mean Absolute Devision Algorithm

```python
data = Generat_datae_with_noises(1000, 10, 0.05)

sb.scatterplot(data)
plt.xticks(rotation = 20)
```

```python
df = pd.DataFrame(data, columns = ["values"])
model = MAD().fit(df)
labels = model.labels_
```

```python
inner_points = data[labels == 0]
outer_points = data[labels == 1]
plt.figure(figsize = (20,5))
sb.scatterplot(inner_points, color = "green")
sb.scatterplot(outer_points, color = "red")
plt.xticks(rotation = 45)
```

## ISolated Forest Algorithm


```python
preds = isolate_forest(pd.Series(data))
plt.figure(figsize = (20,5))
sb.scatterplot(data[preds == 1])
sb.scatterplot(data[preds == -1], color = "red")
```

# Generate Sin-Waved Data

```python
length = 4000
seasonality_amplitude = 10
anomaly_rate = 0.002
anomaly_duration = 10
anomaly_magnitude = 20

sined_data = generate_seasonal_data_with_anomalies(length, seasonality_amplitude, anomaly_rate, anomaly_duration, anomaly_magnitude)
```

```python
plt.figure(figsize = (20,5))
sb.scatterplot(sined_data)
```

## Mean Absolute Devission

```python

labels = mean_abs_devision(sined_data)
pd.Series(labels).unique()
```

```python
plt.figure(figsize = (20, 5))

sb.scatterplot(sined_data[labels == 0])
sb.scatterplot(sined_data[labels == 1])
```

## Isolated Forests

```python
# Create data stream
predictions = isolate_forest(sined_data)
# Train Isolation Forest model

plt.figure(figsize = (20, 5))

plt.scatter(sined_data.index[predictions == -1], sined_data[predictions == -1], color = "red")
plt.scatter(sined_data.index[predictions == 1], sined_data[predictions == 1], color = "green")
plt.xticks(rotation = 45)
```
